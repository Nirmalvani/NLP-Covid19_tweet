### Cleaning text

```python
df
```

```python
def clean_text(text):
    new_text = text.lower() # converting text in lower
    stop = set(stopwords.words("english"))
    new_text = " ".join(word for word in new_text.split() if word.lower() not in stop) # Removing all the stopwords
    new_text = clean(new_text,no_urls=True,no_emoji=True,no_punct=True) # Removing all the punctuation
    lemma = WordNetLemmatizer()
    new_text = " ".join(lemma.lemmatize(word) for word in word_tokenize(new_text)) # Lemmatizing the words
    
    return new_text # Completely cleaned text
```

```python
df['clean_tweet'] = df['tweet'].apply(clean_text)
```

### Model Creation

```python
vector = TfidfVectorizer() # Variable for Tfidvectorizer
```

```python
X = df.iloc[:,1]
y = df.iloc[:,0]
```

```python
print(X.head(2))
print(y.head(2))
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,shuffle=True)
```

```python
tf1_train = vector.fit_transform(X_train)
```

```python
val = pd.read_csv("Constraint_Val.csv")  # Importing validation data
```

```python
val['Label_num'] = val.label.map({"fake":0,"real":1}) # Converting target variable into numerical variable
```

```python
val
```

```python
X_test_val = val.iloc[:,1]
```

```python
y_test_val = val.iloc[:,3]
y_test
```

```python
tf2_test = vector.transform(X_test)
```

```python
def metric(y,pred):
    print(accuracy_score(y,pred))
    print(classification_report(y,pred))
```

```python
lr = LogisticRegression(random_state=42,C=3,max_iter=100,)
```

```python
lr.fit(tf1_train,y_train)
```

```python
pred = lr.predict(tf2_test)
```

```python
metric(y_test,pred)
```

```python
aclr = (round((accuracy_score(y_test,pred))*100))/100
f1lr= (round((f1_score(y_test,pred))*100))/100
```

```python
# Random Forest Classifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(tf1_train,y_train)
```

```python
pred2 = rfc.predict(tf2_test)
```

```python
metric(y_test,pred2)
```

```python
acrfc = (round((accuracy_score(y_test,pred2))*100))/100
f1rfc = (round((f1_score(y_test,pred2))*100))/100
```

```python
# KNearest Neighbours
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(tf1_train,y_train)
```

```python
pred3 = knn.predict(tf2_test)
```

```python
metric(y_test,pred3)
```

```python
acknn = (round((accuracy_score(y_test,pred3))*100))/100
f1knn = (round((f1_score(y_test,pred3))*100))/100
```

```python
tf1_train_ar = tf1_train.toarray()
tf2_test_ar = tf2_test.toarray()
```

```python
tf1_train_ar.shape
```

```python
(tf1_train_ar.shape[0],)
```

```python
# Build a simple neural network model
ann = Sequential([
    Dense(128, activation='sigmoid', input_shape=(tf1_train_ar.shape[1],)),
    Dropout(0.1),
    Dense(64, activation='sigmoid'),
    Dense(1, activation='sigmoid')
])

# Compile the model
ann.compile(optimizer=Adam(learning_rate=0.01),
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
# Train the model
history = ann.fit(tf1_train_ar, y_train, epochs=2, batch_size=32, validation_split=0.3)
```

```python
pred4 = ann.predict(tf2_test_ar)
```

```python
y_pred_new = (pred4>=0.5)
```

```python
metric(y_test,y_pred_new)
```

```python
acann = (round((accuracy_score(y_test,y_pred_new))*100))/100
f1ann = (round((f1_score(y_test,y_pred_new))*100))/100
```

```python
from prettytable import PrettyTable
```

```python
table = PrettyTable()
table.field_names = ['Model','Accuracy','F1 score']
```

```python
lr, rfc, knn ,ann
```

```python
model_data = [
    ['Logistic Regression',aclr,f1lr],
    ['Random Forest Classifier',acrfc,f1rfc],
    ['KNearest Neighbours',acknn,f1knn],
    ['Ann',acann,f1ann]
]

for row in model_data:
    table.add_row(row)
```
